```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/output.csv")
df = df[df["experimento"] == "problema_2"]
df
```

```python
df_plot = df.groupby("algoritmo")["tempo_execucao"].agg(["mean", "std"]).reset_index()
df_plot
```

```python
bars = plt.bar(
  df_plot["algoritmo"], 
  df_plot["mean"], 
  yerr=df_plot["std"], 
  capsize=10, 
  color=["#5DADE2", "#EC7063"],
  edgecolor="black",
)

plt.ylabel("Tempo médio de 10 execuções (s)", fontsize=12)
# plt.yscale("log")
plt.grid(axis="y", linestyle="--", alpha=0.3)

for bar, std_val in zip(bars, df_plot["std"]):
    y_val = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, 
             y_val + std_val, 
             f"{std_val:.5f} s", 
             ha="center", 
             va="bottom", 
             fontsize=10)

plt.tight_layout()
# Armazenamos o gráfico em formato pdf para utilizá-lo no documento LaTex.
plt.savefig("../data/grafico_barras_problema_2.pdf", dpi=300, bbox_inches="tight")
# Também armazenamos em formato png para facilitar a visualização pelo repositório no GitHub.
plt.savefig("../data/grafico_barras_problema_2.png", dpi=300, bbox_inches="tight")
plt.show()
```
